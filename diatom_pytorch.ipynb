```python
import os
import numpy as np
import pandas as pd
import cv2
import tifffile
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.models as models
from torchvision import transforms
from pathlib import Path
import random
from tqdm import tqdm

# Set random seeds EXACTLY as in original code
np.random.seed(42)
torch.manual_seed(42)
random.seed(42)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

```python
class DiatomCountModel(nn.Module):
    """
    PyTorch model for diatom counting using transfer learning.
    EXACTLY the same architecture as in the original code.
    """
    def __init__(self):
        super(DiatomCountModel, self).__init__()
        
        # Use pre-trained ResNet18 as base model
        self.base_model = models.resnet18(pretrained=True)
        
        # Replace final layer with a regression head
        num_features = self.base_model.fc.in_features
        self.base_model.fc = nn.Identity()  # Remove classifier
        
        # Create regression head
        self.regression_head = nn.Sequential(
            nn.Linear(num_features, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 1),
            nn.ReLU()  # ReLU ensures non-negative count
        )
    
    def forward(self, x):
        features = self.base_model(x)
        count = self.regression_head(features)
        return count


class DiatomDataset(Dataset):
    """
    PyTorch Dataset for a single image, using EXACT same preprocessing steps.
    """
    def __init__(self, image, transform=None):
        self.image = image
        self.transform = transform
    
    def __len__(self):
        return 1
    
    def __getitem__(self, idx):
        image = self.image
        
        # Convert to PyTorch tensor
        image = torch.from_numpy(image).permute(2, 0, 1).float()  # (H,W,C) -> (C,H,W)
        
        # Apply transforms if any
        if self.transform:
            image = self.transform(image)
        
        # Return with dummy label which will be ignored during prediction
        return image, torch.tensor(0.0, dtype=torch.float32)


def predict_and_visualize_single_image(image_path, model_path, csv_path, output_path=None,
                                      target_size=(512, 512), 
                                      device='cuda' if torch.cuda.is_available() else 'cpu'):
    """
    Load a single image, predict diatom count using a trained model, and visualize the result
    using EXACTLY the same preprocessing and prediction pipeline as the original code.
    
    Args:
        image_path (str): Path to the image file (TIFF format)
        model_path (str): Path to the saved model state dict
        csv_path (str): Path to the CSV file containing ground truth data
        output_path (str, optional): Path to save the visualization
        target_size (tuple): Size to resize the image to (height, width)
        device (str): Device to run the model on
    
    Returns:
        fig: Matplotlib figure with visualization
        prediction: The predicted count (float)
        true_count: The true count from the CSV (or None if not found)
    """
    # Step 1: Get the image ID from the path
    image_name = os.path.basename(image_path)
    image_id = os.path.splitext(image_name)[0]  # Remove extension
    print(f"Processing image: {image_id}")
    
    # Step 2: Load the CSV and find the true count for this image
    true_count = None
    try:
        df = pd.read_csv(csv_path)
        # Look for the image ID in the micrograph ID column
        matching_row = df[df['micrograph ID'] == image_id]
        if not matching_row.empty:
            true_count = matching_row['Cocconeis'].values[0]
            print(f"Found true count for {image_id}: {true_count}")
        else:
            print(f"Warning: Could not find {image_id} in the CSV file")
    except Exception as e:
        print(f"Error loading CSV file: {e}")
    
    # Step 3: Load and preprocess the image EXACTLY as in the DiatomDataset.preprocess_image
    # This is critical for consistent results
    image = tifffile.imread(image_path)
    print(f"Original image shape: {image.shape}")
    
    # Resize image
    image = cv2.resize(image, target_size)
    
    # Normalize to [0, 1]
    image = image.astype(np.float32) / 255.0
    print(f"Preprocessed image shape: {image.shape}")
    
    # Save a copy for visualization
    image_display = image.copy()
    
    # Step 4: Create dataset and dataloader with EXACT same normalization as in original code
    # The normalization values MUST match those used during training
    transform = transforms.Normalize(
        mean=[0.485, 0.456, 0.406], 
        std=[0.229, 0.224, 0.225]
    )
    
    # Create dataset
    dataset = DiatomDataset(image, transform)
    
    # Create dataloader with batch size 1
    dataloader = DataLoader(
        dataset,
        batch_size=1,
        shuffle=False,
        num_workers=0  # Use 0 for simplicity
    )
    
    # Step 5: Load and prepare the model
    # Load model architecture
    model = DiatomCountModel()
    
    # Load weights
    print(f"Loading model from {model_path}")
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    model.eval()  # Set to evaluation mode
    
    # Step 6: Make prediction
    with torch.no_grad():
        for images, _ in dataloader:
            images = images.to(device)
            outputs = model(images)
            prediction = outputs.cpu().numpy().item()
            prediction_rounded = round(prediction)
    
    print(f"Raw prediction: {prediction}, Rounded: {prediction_rounded}")
    
    # Step 7: Create visualization with the EXACT same style as the original
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Display the image
    ax.imshow(image_display)
    
    # Calculate error metrics if true count is available
    if true_count is not None:
        error = prediction_rounded - true_count
        
        # Calculate percentage error
        if true_count != 0:
            pct_error = (abs(error) / true_count) * 100
        else:
            pct_error = float('inf')  # Handle division by zero
        
        # Set title with count information - MATCH EXACTLY the original format
        title = f"True: {true_count}, Predicted: {prediction_rounded}, Error: {error}"
        
        # Add percentage error in red text on the image
        ax.text(10, 30, f"Error: {error} ({pct_error:.2f}%)", color='red', fontsize=12, 
                bbox=dict(facecolor='white', alpha=0.7))
    else:
        title = f"Predicted: {prediction_rounded}"
    
    ax.set_title(title)
    ax.axis('off')
    plt.tight_layout()
    
    # Save visualization if output path is provided
    if output_path:
        plt.savefig(output_path)
        print(f"Saved visualization to {output_path}")
    
    return fig, prediction, true_count


def reproduce_prediction_visualization(model_path, test_dir, csv_path, output_dir, 
                                      num_samples=5, device='cuda' if torch.cuda.is_available() else 'cpu'):
    """
    Reproduce the format of the original prediction_visualization.png by creating
    a vertical stack of prediction images.
    
    Args:
        model_path (str): Path to the model
        test_dir (str): Directory containing test images
        csv_path (str): Path to CSV with ground truth
        output_dir (str): Directory to save visualizations
        num_samples (int): Number of images to visualize
        device (str): Device to run on
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Get all TIFF files in the test directory
    image_files = list(Path(test_dir).glob('*.tif'))
    
    # Select a subset if requested
    if num_samples < len(image_files):
        # Set seed for reproducibility
        random.seed(42)
        image_files = random.sample(image_files, num_samples)
    
    # Create a figure to match the original visualization format
    # Each image will be in its own subplot, stacked vertically
    fig, axes = plt.subplots(num_samples, 1, figsize=(8, 8*num_samples))
    if num_samples == 1:
        axes = [axes]  # Make sure axes is a list for iteration
    
    # Process each image
    for i, image_file in enumerate(image_files):
        print(f"\nProcessing image {i+1}/{len(image_files)}: {image_file.name}")
        
        # Get image ID
        image_id = image_file.stem
        
        # Load CSV and find true count
        true_count = None
        try:
            df = pd.read_csv(csv_path)
            matching_row = df[df['micrograph ID'] == image_id]
            if not matching_row.empty:
                true_count = matching_row['Cocconeis'].values[0]
        except Exception as e:
            print(f"Error loading CSV file: {e}")
        
        # Load and preprocess image
        image = tifffile.imread(str(image_file))
        image = cv2.resize(image, (512, 512))
        image = image.astype(np.float32) / 255.0
        
        # Create dataset with normalization
        transform = transforms.Normalize(
            mean=[0.485, 0.456, 0.406], 
            std=[0.229, 0.224, 0.225]
        )
        dataset = DiatomDataset(image, transform)
        dataloader = DataLoader(dataset, batch_size=1, shuffle=False)
        
        # Load model and make prediction
        model = DiatomCountModel()
        model.load_state_dict(torch.load(model_path, map_location=device))
        model.to(device)
        model.eval()
        
        with torch.no_grad():
            for images, _ in dataloader:
                images = images.to(device)
                outputs = model(images)
                prediction = outputs.cpu().numpy().item()
                prediction_rounded = round(prediction)
        
        # Display image in the current subplot
        axes[i].imshow(image)
        
        # Set title in EXACTLY the same format as the original
        if true_count is not None:
            error = prediction_rounded - true_count
            title = f"True: {true_count}, Predicted: {prediction_rounded}, Error: {error}"
        else:
            title = f"Predicted: {prediction_rounded}"
        
        axes[i].set_title(title)
        axes[i].axis('off')
    
    # Adjust layout and save
    plt.tight_layout()
    output_path = os.path.join(output_dir, "prediction_visualization_reproduced.png")
    plt.savefig(output_path)
    plt.close(fig)
    print(f"Saved combined visualization to {output_path}")


def batch_predict_single_style(image_dir, model_path, csv_path, output_dir, 
                              num_samples=None, random_selection=True, 
                              device='cuda' if torch.cuda.is_available() else 'cpu'):
    """
    Process multiple images one at a time (not in batch) to produce individual visualizations
    in the style of the original prediction_visualization.png.
    
    Args:
        image_dir (str): Directory containing TIFF images
        model_path (str): Path to the saved model state dict
        csv_path (str): Path to the CSV file containing ground truth data
        output_dir (str): Directory to save visualizations
        num_samples (int, optional): Number of images to process. If None, process all images.
        random_selection (bool): If True and num_samples is set, select images randomly
        device (str): Device to run the model on
    
    Returns:
        results_df: DataFrame with prediction results
    """
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    # Get all TIFF files in the directory
    image_files = list(Path(image_dir).glob('*.tif'))
    
    # If num_samples is specified, select that many images
    if num_samples is not None and num_samples < len(image_files):
        if random_selection:
            # Set seed for reproducibility
            random.seed(42)
            image_files = random.sample(image_files, num_samples)
        else:
            image_files = image_files[:num_samples]
    
    # Store results
    results = []
    
    # Process each image
    for i, image_file in enumerate(tqdm(image_files, desc="Processing images")):
        print(f"\nProcessing image {i+1}/{len(image_files)}: {image_file.name}")
        
        # Set output path for this image
        output_path = os.path.join(output_dir, f"{image_file.stem}_prediction.png")
        
        # Make prediction and visualize
        _, prediction, true_count = predict_and_visualize_single_image(
            image_path=str(image_file),
            model_path=model_path,
            csv_path=csv_path,
            output_path=output_path,
            device=device
        )
        
        # Calculate error metrics if true count is available
        absolute_error = None
        percentage_error = None
        
        if true_count is not None:
            absolute_error = round(prediction) - true_count
            if true_count != 0:
                percentage_error = (abs(absolute_error) / true_count) * 100
        
        # Store results
        results.append({
            'image_id': image_file.stem,
            'true_count': true_count,
            'predicted_count': prediction,
            'predicted_count_rounded': round(prediction),
            'absolute_error': absolute_error,
            'percentage_error': percentage_error
        })
    
    # Create and return results DataFrame
    results_df = pd.DataFrame(results)
    
    # Save results to CSV
    results_csv_path = os.path.join(output_dir, "prediction_results.csv")
    results_df.to_csv(results_csv_path, index=False)
    print(f"Results saved to {results_csv_path}")
    
    # Calculate and print overall metrics
    if 'absolute_error' in results_df and not results_df['absolute_error'].isna().all():
        mae = results_df['absolute_error'].abs().mean()
        mse = (results_df['absolute_error'] ** 2).mean()
        rmse = np.sqrt(mse)
        
        print("\nOverall Metrics:")
        print(f"Mean Absolute Error: {mae:.2f}")
        print(f"Mean Squared Error: {mse:.2f}")
        print(f"Root Mean Squared Error: {rmse:.2f}")
        
        # Calculate mean percentage error (excluding cases where true_count is 0)
        valid_pct_errors = results_df['percentage_error'].dropna()
        if len(valid_pct_errors) > 0:
            mape = valid_pct_errors.mean()
            print(f"Mean Absolute Percentage Error: {mape:.2f}%")
    
    return results_df


# Example usage:
if __name__ == "__main__":
    # Set paths (using the paths from the original code)
    csv_path = "/projects/genomic-ml/da2343/diatom/Kraken_2023_measurements.csv"
    model_path = "/projects/genomic-ml/da2343/diatom/models/diatom_counter_best.pth"
    test_dir = "/projects/genomic-ml/da2343/diatom/test/"
    output_dir = "/projects/genomic-ml/da2343/diatom/predictions/"
    
    # Force CPU to match original prediction environment if needed
    # Comment out the line below if you want to use CUDA when available
    device = 'cpu'  # Force CPU to match original environment
    print(f"Using device: {device}")
    
    # Option 1: Single image prediction with the exact same visualization style
    # If you have a specific image you want to test, specify its path here
    specific_image_path = None  # e.g., "/projects/genomic-ml/da2343/diatom/test/specific_image.tif"
    
    if specific_image_path and os.path.exists(specific_image_path):
        # Use the specific image
        image_path = specific_image_path
    else:
        # Choose the first image from the test directory
        image_files = list(Path(test_dir).glob('*.tif'))
        if image_files:
            image_path = str(image_files[0])
        else:
            print("No test images found.")
            image_path = None
    
    if image_path:
        print("\n*** Single image prediction ***")
        fig, prediction, true_count = predict_and_visualize_single_image(
            image_path=image_path,
            model_path=model_path,
            csv_path=csv_path,
            output_path=os.path.join(output_dir, "single_prediction_result.png"),
            device=device
        )
        
        plt.close(fig)
        
        print(f"Predicted diatom count: {prediction:.2f} (rounded: {round(prediction)})")
        if true_count is not None:
            print(f"True count: {true_count}")
            error = round(prediction) - true_count
            print(f"Error: {error}")
            if true_count != 0:
                print(f"Percentage error: {(abs(error) / true_count) * 100:.2f}%")
    
    # Option 2: Reproduce the original prediction_visualization.png format with multiple images
    print("\n*** Reproducing prediction visualization with multiple images ***")
    reproduce_prediction_visualization(
        model_path=model_path,
        test_dir=test_dir,
        csv_path=csv_path,
        output_dir=output_dir,
        num_samples=5,  # Adjust to match the number of images in the original visualization
        device=device
    )
    
    # Option 3: Process multiple images one at a time
    print("\n*** Processing multiple images one at a time ***")
    results_df = batch_predict_single_style(
        image_dir=test_dir,
        model_path=model_path,
        csv_path=csv_path,
        output_dir=output_dir,
        num_samples=5,  # Process 5 images
        random_selection=True,  # Select randomly
        device=device
    )
    
    print("Processing complete!")
```
